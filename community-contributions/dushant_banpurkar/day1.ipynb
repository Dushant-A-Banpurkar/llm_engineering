```python
info=fetch_website_contents("https://github.com/uber/h3")

print(info)
```

```python
system_prompt=""""
Your a smart ai research assistant that analyize the github repositories and content of the repositories,
break down each files and provide research, how the accomplish on there agenda and how the implement there idea into a smart reality.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

```python
user_prompt="""
Here is the the github repositories.
Provide summary for this github repositories.
If it include major changes or important files summarize them too.
"""
```

```python
message=[
    {"role":"system","content":"Your a live ipl match assistant"}
    ,{"role":"user","content":"Which maping techinique is used by UBER to map rider?"}
]

response=ollama.chat.completions.create(model="llama3.2",messages=message)
response.choices[0].message.content
```

```python
def message_for(repositorie):
    return [
        {"role":"system","content":system_prompt},
        {"role":"user","content":user_prompt+repositorie}
    ]
    
```

```python
message_for(info)
```

```python
def summarize(url):
    respositeries =fetch_website_contents(url)
    response =ollama.chat.completions.create(
        model="llama3.2",
        messages=message_for(respositeries)
    )
    return response.choices[0].message.content
```

```python
summarize("https://github.com/uber/h3")
```

```python
def display_summary(url):
    summary=summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://github.com/uber/h3")
```
